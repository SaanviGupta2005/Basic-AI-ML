# BIVARIANT ANALSIS

# HOW HAS BETTER CANCE OF SURVIVAL MALE OR FEMALE

```python
# Count the number of male and female passengers who survived
male_survived_count = df[(df[1] == 1) & (df[4] == 'male')][0].count()
female_survived_count = df[(df[1] == 1) & (df[4] == 'female')][0].count()

# Calculate total number of male and female passengers
total_male_count = df[df[4] == 'male'][0].count()
total_female_count = df[df[4] == 'female'][0].count()

# Calculate survival rates
male_survival_rate = male_survived_count / total_male_count
female_survival_rate = female_survived_count / total_female_count

print("Male survival rate:", male_survival_rate)
print("Female survival rate:", female_survival_rate)
```

# WHICH PASSENGER CLASS HAS BETTER CHANCE OF SURVIVAL(FIRST,SECOND,OR THIRD CLASS)?

```python
# Count the number of male and female passengers
third_survived_count = df[(df[1] == 1) & (df[2] == 3)][0].count()
second_survived_count = df[(df[1] == 1) & (df[2] == 2)][0].count()
first_survived_count = df[(df[1] == 1) & (df[2] == 1)][0].count()

# Calculate total number of male and female passengers
total_third_count = df[df[2] == 3][0].count()
total_second_count = df[df[2] == 2][0].count()
total_first_count = df[df[2] == 1][0].count()

# Calculate survival rates
third_survival_rate = third_survived_count / total_third_count
second_survival_rate = second_survived_count / total_second_count
first_survival_rate = first_survived_count / total_first_count

print("First survival rate:", first_survival_rate)
print("Second survival rate:", second_survival_rate)
print("Third survival rate:", third_survival_rate)
```
